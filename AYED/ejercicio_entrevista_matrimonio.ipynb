# Especificación

Para poder casarte con el amor de tu vida, es necesario poder contestar una pregunta
relacionada con el siguiente código:

Como se puede apreciar, existe una función llamada trib, la cual se llama con dos enteros
como parámetros. Si prueba el código anterior, se podrá dar cuenta que la función trib()
es llamada un total de 41 veces cuando es llamada desde la función principal como
trib(5, 5). La idea es determinar el número de veces que la función trib será invocada
para diferentes valores de n y back.

# Entrada

El archivo de entrada tiene varias líneas de entrada Cada línea contiene 2 enteros n(n ≤
61) y back(back ≤ 60).


# Salida

El archivo de salida tiene varias líneas de respuesta. Cada línea contiene el número de
caso en el formato Case < numero > y la cantidad de veces que la función trib será
invocada para diferentes valores de n y back.


# Diseño

---



# Estrategia

Lo que haremos será una función recursiva que se ejecutará hasta que n sea menor o igual a 0, e iremos contando las veces que se ejecuta la recursión, este contador es el que entregaremos como resultado.

# Implementación del Algoritmo



---


```python
def trib(n, back):
    global cont
    cont+=1
    if n<=0:
        return 0
    if n == 1:
        return 1
    acum = 0
    for i in range(1, back+1):
        acum += trib(n-i, back)
    return acum

def main():
    global cont
    cont=0
    t=input("Ingrese: ")
    y=t.split()
    n=y[0]
    back=y[1]
    trib(int(n), int(back))
    print("Para el caso",n,back,"la respuesta es:",cont)
    main()
main()
```
