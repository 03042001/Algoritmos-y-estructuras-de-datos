# Especificación

Jesse ama las galletas y quiere que sus galletas tengan una dulzura mayor a un valor k; para hacer esto combina las dos galletas que tengan la menor dulzura, esto crea una galleta con una dulzura de (1*galleta con menor dulzura + 2*segunda galleta con menor dulzura) y repite este proceso hasta que todas las galletas tengan una dulzura mayor a k. Determinar el numero de operaciones minimas para realizar este proceso.

# Entrada

La primera linea son dos enteros separados por espacio n y k, que son el tamaño del arrgelo y la dulzura minimma requerida.

# Salida

Un entero que es el numero de iteraciones requeridas.

# Estrategia
Lo que haremos será tomar la estructura de colas y modificarla como una cola de prioridad minima, es decir que tendremos que agregar funciones como extraer el minimo y eliminar el minimo, en el constructor inicializaremos la cola con la lista de galletas con la que se hará el ejercicio; una vez hecho esto crearemos una función solucion que se encargará de realizar lo que pide el ejercicio, en este lo que haremos será crear una variable que guardará el resultado, mediante un ciclo extraeremos los dos minimos y los guardaremos en variables para despues combinarlos como nos pide el ejercicio, despues de extraerlos los eliminamos y la mezcla entre ambos la encolamos a la lista original, lo unico que falta es sumarle uno a la variable que dará el resultado, este proceso se repetirá hasta que todas las galletas tengan una dulzura mayor a la que nos piden.

# Estructura de datos

Para este ejercicio usaremos una cola de prioridad minima como estructura de datos, esto nos facilitará la solución del ejercicio debido a que la base del ejercicio es extraer los valores minimos de una lista y esta es una de las principales funciones de una cola de prioridad minima.

# Implementación del Algoritmo



---


```python
class jesse_cookies:

    def __init__(self, galletas):
        self.queue = galletas

    def extraer_minimo(self):
        return min(self.queue)

    def eliminar_minimo(self):
        minimo = min(self.queue)
        minimo_index = self.queue.index(minimo)
        self.queue.pop(minimo_index)

    def solucion(self, valor):
            respuesta = 0
            while self.extraer_minimo() < valor:
                minimo1 = self.extraer_minimo()
                self.eliminar_minimo()

                minimo2 = self.extraer_minimo()
                self.eliminar_minimo()

                self.enqueue((1 * minimo1) + (2 * minimo2))
                respuesta += 1

            return respuesta

    def size_queue(self):
        return len(self.queue)

    def enqueue(self, value):
        self.queue.insert(0, value)

    def queue_is_empty(self):
        return self.size_queue() == 0
```
