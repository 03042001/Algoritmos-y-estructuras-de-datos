# Examen tercio 1, punto 4
Ejercicio: "Closest numbers"

# Requisitos

---



# Especificación

Tenemos que determinar que pares de elementos tienen la diferencia minima absoluta entre ellos, si hay muchos pares encontrarlos todos.

# Entrada

La primera linea es un entero n, la longitud del arreglo, la segunda linea contiene n enteros separados por espacio.

# Salida

Imprima el numero de posibles combinaciones para secuencias de numeros naturales unicos elevados a la enesima potencia que suman x.

# Diseño

---



# Estrategia
Lo primero que haremos es ordenar el arreglo, esto nos facilitará el trabajo cuando tengamos que calcular la diferencia entre los distintos elementos del arreglo, una vez ordenado crearemos una lista que guardará todas las diferencias (restas) entre los elementos de la lista, estas diferencias las calcularemos iterativamente mediante un ciclo que va restando el siguiente elemento con el actual (esto lo hacemos iterativamente mediante un ciclo), a esa lista de diferencias le sacamos el minimo, esto nos va a dar el entero que corresponde a la menor diferencia que hay entre los elementos, lo unico que faltaria es ir recorriendo la lista de diferencias e ir comparando si la diferencia actual es igual a la diferencia minima, en caso positivo agragamos a una lista resultado dos valores que son los que restados dan esa diferencia minima.

# Estructura de datos

Para solucionar este ejercicio usaremos listas.

# Implementación del Algoritmo



---


```python
def closest(arr):
    diff = []
    arr = sorted(arr)
    for i in range(len(arr) - 1):
        diff.append(arr[i+1] - arr[i])
    min_diff = min(diff)
    respuesta = []
    for i in range(len(diff)):
        if diff[i] == min_diff:
            respuesta.append(arr[i])
            respuesta.append(arr[i+1])
    return respuesta

def main():
    n = int(input().strip())
    arr = list(map(int, input().rstrip().split()))
    result = closest(arr)
    print(*result)
main()
```

# Invariante

En este algoritmo el arreglo de entrada siempre tendrá una longitud mayor que 1, esto se debe a que si el arreglo de entrada es 1 habrá un error a la hora de calcular la diferencia minima. Además esta propiedad se cumple tanto en el inicio como en la terminación del algoritmo.
