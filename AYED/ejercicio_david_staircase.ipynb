# Especificación

Davis tiene cierto número de escaleras de distintos tamaños en su casa, él puede subir de 1 a 3 escalones y quiere calcular cuantas maneras hay de subir hasta el tope de cada escalera.

# Entrada

La primera linea contiene un entero "s", que es el número de escaleras que hay en la casa de Davis, cada una de las siguientes "s" lineas equivale al número de escalones que tiene cada escalera.

# Salida


Un número "s" de lineas, cada una con un entero que representa de maneras que se puede cada escalera.

# Estrategia

Para solucionar este ejercicio recursivo lo analizaremos desde lo general a lo individual:
Pensemos que hay un número n de escalones, para llegar al ultimo escalón, el escalón n, David puede llegar de a dos en dos (n-2), de a uno en uno (n-1) o de a tres en tres (n-3), si sumamos el número de formas en que puede subir de estas tres maneras ya tendriamos el resultado deseado, ahora pensemos en los casos base, si al subir llegó al ultimo escalón quiere decir que es una forma valida, es decir que retornaría 1, mientras que si al subir se "pasa" no sería una forma valida y retornaria 0.

# Implementación del Algoritmo



---


```python
def escalera(n, dinamica = {}):
    if n in dinamica:
      return dinamica[n]
    if  n < 0:
        return 0
    elif n==0:
        return 1
    else:
        respuesta = escalera(n-1, dinamica)+ escalera(n-2, dinamica)+escalera(n-3, dinamica)
        dinamica[n] = respuesta
        return respuesta

def main():
    s = int(input().strip())
    for s_itr in range(s):
        n = int(input().strip())
        res = escalera(n)

    print(res)

main()
```

# Invariante

En este algoritmo podemos observar que tanto n (numero de escalones) como el numero de formas (respuesta) siempre va a ser un número mayor o igual que 0.

1.   Iniciación: El número n es un entero igual o mayor a 0.
2.   Estabilidad: Durante el proceso el resultado del algoritmo no será negativo.
3. Terminación: Cuando llegue a su recursión maxima (casos base), el resultado permanecerá mayor que 0.
