```python
#Punto a.

def generar_sucesion(n):
    sucesion = [1, 1]
    for i in range(2, n):
        sucesion.append(sucesion[i-1] + sucesion[i-2])
    return sucesion

terminos_sucesion = generar_sucesion(50)
print("Términos de la sucesión:")
print(terminos_sucesion)
```

```python
#Punto c.

def generar_sucesion(n):
    sucesion = [1, 1]
    for i in range(2, n):
        sucesion.append(sucesion[i-1] + sucesion[i-2])
    return sucesion

def construir_sucesion_ratio(sucesion):
    sucesion_ratio = [sucesion[i+1] / sucesion[i] for i in range(len(sucesion)-1)]
    return sucesion_ratio

terminos_sucesion = generar_sucesion(50)
sucesion_ratio = construir_sucesion_ratio(terminos_sucesion)
print("Términos de la sucesión de ratios:")
print(sucesion_ratio)
```

```python
#Punto e.

from sympy import symbols, Eq, solve

x = symbols('x')

ecuacion = Eq(1/x, x - 1)
soluciones = solve(ecuacion, x)
print("Soluciones de la ecuación:")
for solucion in soluciones:
    print(f"x = {solucion}")
```

```python
# Punto g.

razones_largo_ancho = {
    "Tarjeta débito": 85.6 / 53.98,  # Dimensiones estándar de una tarjeta de débito
    "Tarjeta de crédito": 85.6 / 53.98,  # Dimensiones estándar de una tarjeta de crédito
    "Nueva cédula digital": 86 / 54,  # Dimensiones estándar de una cédula digital
    "Dólar americano": 155.956 / 66.294,  # Dimensiones estándar de un billete de dólar americano
    "Dólar canadiense": 152.4 / 69.85,  # Dimensiones estándar de un billete de dólar canadiense
    "Billete de 50 euros": 140 / 77,  # Dimensiones estándar de un billete de 50 euros
    "Billete de 500 euros": 160 / 82,  # Dimensiones estándar de un billete de 500 euros
    "Billete de 2000 pesos colombianos": 140 / 70,  # Dimensiones aproximadas de un billete de 2000 pesos colombianos
    "Pantalla de computadora": 16 / 9  # Relación de aspecto estándar de una pantalla de computadora
}
razones_ordenadas = sorted(razones_largo_ancho.values())
print("\nLista de razones largo y ancho por objeto:\n")
for objeto, razon in razones_largo_ancho.items():
    print(f"{objeto}: {razon}")

print("\nLista ordenada de razones de largo y ancho:\n")
for objeto in razones_ordenadas:
    print(objeto)
```
