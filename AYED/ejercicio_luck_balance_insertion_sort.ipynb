# Especificación

Lena se esté preparando para una importante competición de código que está precedida por unos concursos premilinares; ella cree en "acumular suerte" y quiere probar su teoría, cada concurso se describe por dos enteros, L[i] y T[i], L[i] es la cantidad de suerte acumulada y T[i] es la importancia del concurso; nos toca calcular el máximo equilibrio de suerte alcanzable

# Entrada

Dos enteros separados por espacio n y k, el número de concursos preliminares y el máximo número de concursos importantes que Lena puede perder. Cada una de las siguientes n lineas contienen dos enteros separados por espacio L{i] y T[i], el equilibrio de suerte del concurso y su clasificación de importancia.

# Salida

Un entero que equivale al máximo equilibrio de suerte alcanzable.

# Diseño

---



# Estrategia
Para la solución de este ejercicio vamos a pensar en las listas que vamos a tomar para solucionarlo, lo primero que hacemos es hacer una lista de los concursos importantes, es decir tomar las tuplas cuyo segundo elemento es 1, procedemos a ordenarlos pero no ascendentemente sino descendentemente, esto debido a que estamos perdiendo primero los concursos más valiosos en términos de suerte; después calculamos la suma total de suerte de todos los concursos sin importar su importancia, después eliminamos la suerte que se pierde dependiendo de los concursos ganados y retornamos el total de suerte acumulada.

# Estructura de datos
Para el desarrollo de este ejercicio usaremos principalmente listas.


# Implementación del Algoritmo



---



```python
def insertion_sort(arr):
    """Método de organización insertion sort
    list -> list
    """
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

def luckBalance(k, contests):
    important_contests = [(luck, importance) for luck, importance in contests if importance == 1]
    insertion_sort(important_contests)

    total_luck = sum(luck for luck, _ in contests)
    for i in range(len(important_contests) - k):
        total_luck -= important_contests[i][0] * 2

    return total_luck

def main():
    first_multiple_input = input().rstrip().split()
    n = int(first_multiple_input[0])
    k = int(first_multiple_input[1])
    contests = []
    for _ in range(n):
        contests.append(list(map(int, input().rstrip().split())))

    result = luckBalance(k, contests)
    return result

main()
```
