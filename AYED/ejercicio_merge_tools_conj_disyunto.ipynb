```python
class conjuntoDisyunto:
    def __init__(self, elementos):
        self.parent = {el: el for el in elementos}
        self.rank = {el: 0 for el in elementos}

    def buscar(self, elemento):
        if self.parent[elemento] != elemento:
            self.parent[elemento] = self.find(self.parent[elemento])
        return self.parent[elemento]

    def union(self, elemento1, elemento2):
        root1 = self.buscar(elemento1)
        root2 = self.buscar(elemento2)

        if root1 != root2:
            if self.rank[root1] < self.rank[root2]:
                self.parent[root1] = root2
            elif self.rank[root1] > self.rank[root2]:
                self.parent[root2] = root1
            else:
                self.parent[root1] = root2
                self.rank[root2] += 1

def proceso_substrings(s, k):
    n = len(s)
    substrings = [s[i:i+k] for i in range(0, n, k)]
    lista = []

    for substring in substrings:
        elements = list(set(substring))
        ds = conjuntoDisyunto(elements)
        lista.append(ds)

    for i in range(1, len(lista)):
        anterior = lista[i - 1]
        actual = lista[i]

        for element in actual.parent:
            if element in anterior.parent:
                anterior.union(element, anterior.buscar(element))

    result = [''.join(sorted(set(substring), key=s.index)) for substring in substrings]

    for res in result:
        print(res)
```
