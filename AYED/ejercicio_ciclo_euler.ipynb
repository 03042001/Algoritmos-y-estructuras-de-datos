# Estrategia

Para este ejercicio reutilizaremos el codigo que usamos en el ejercicio 1, en este tenemos un grafo cuyos vertices son diccionarios y ademas de las funciones normales de un grafo tendremos una busqueda por profundidad y una funcion para encontrar los componentes que tienen conexiones.

Lo nuevo que implementaremos para resolver este ejercicio será una función que irá recorriendo el grafo y verificará si cada vertice tiene un número de conexiones impar, en este caso retornará False y en la función main se dará un No como respuesta, en caso de que no encuentre ningun vertice con un numero de conexiones impar se entenderá que se cumple la condicion y por lo tanto en la funcion main se dará un Si.

# Estructura de datos

Usaremos un grafo no dirigido como estructura de datos.

# Implementación del Algoritmo



---


```python
class Grafo:
    def __init__(self):
        self.vertices = {}

    def agregar_vertice(self, vertice):
        if vertice not in self.vertices:
            self.vertices[vertice] = []

    def agregar_arista(self, vertice1, vertice2):
        self.agregar_vertice(vertice1)
        self.agregar_vertice(vertice2)

        self.vertices[vertice1].append(vertice2)
        self.vertices[vertice2].append(vertice1)

    def dfs(self, inicio, visitado, componente_actual):
        visitado[inicio] = True
        componente_actual[0] += 1

        for vecino in self.vertices[inicio]:
            if not visitado[vecino]:
                self.dfs(vecino, visitado, componente_actual)

    def encontrar_componentes_conexos(self):
        visitado = {vertice: False for vertice in self.vertices}
        tamaños_componentes = []

        for vertice in self.vertices:
            if not visitado[vertice]:
                componente_actual = [0]
                self.dfs(vertice, visitado, componente_actual)
                tamaños_componentes.append(componente_actual[0])

        return tamaños_componentes

    def euler(self):
        tamaños_componentes = self.encontrar_componentes_conexos()

        for vertice in self.vertices:
            if len(self.vertices[vertice]) % 2 != 0:
                return False
        return len(tamaños_componentes) == 1

def main():
    grafo = Grafo()
    r = int(input(""))
    for i in range(r):
        t = input("")
        p = t.split()
        grafo.agregar_arista(int(p[0]), int(p[1]))

    if grafo.euler():
        print("Sí")
    else:
        print("No")

if __name__ == "__main__":
    main()
```
