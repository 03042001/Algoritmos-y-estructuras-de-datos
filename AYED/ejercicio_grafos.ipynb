# Especificación

Existen 2N posibles números enteros para representar los nodos (vértices) en un grafo G
no dirigido. Dicho grafo consta de N arcos (A1, B1),(A2, B2). . . ,(AN , BN ) de tal manera
que Ai ∈ [1, N] y Bi ∈ [N + 1, 2N] para i(1 ≤ i ≤ N).
Su tarea será encontrar el número de vértices en el componente conexo más pequeño y
el número de vértices en el más grande.

# Entrada

La entrada constará de un único caso de prueba. La primera línea de la entrada contiene
un entero N(1 ≤ N ≤ 1500) que indica el número de arcos de G, Las N siguientes líneas
tendrán dos enteros positivos Ai(1 ≤ Ai ≤ N) y Bi(N + 1, ≤ Bi ≤ 2N) que representan
el arco (Ai
, Bi) de G.

# Salida
Imprima dos enteros positivos min max separados por un espacio indicando min el
número de nodos en cualquiera de los componentes conexos más grandes.


# Diseño

---



# Estrategia

Lo que haremos será ir creando un grafo a través de las conexiones dadas y luego usaremos DFS para explorar y calcular los tamaños de los componentes conexos en el grafo, guardaremos los vértices que ya fueron visitados en un diccionario.

# Estructura de datos
Para este ejercicio usaremos un grafo como estructura de datos.

# Implementación del Algoritmo



---


```python
class Grafo:
    def __init__(self):
        self.vertices = {}

    def agregar_vertice(self, vertice):
        if vertice not in self.vertices:
            self.vertices[vertice] = []

    def agregar_arista(self, vertice1, vertice2):
        self.agregar_vertice(vertice1)
        self.agregar_vertice(vertice2)

        self.vertices[vertice1].append(vertice2)
        self.vertices[vertice2].append(vertice1)

    def dfs(self, inicio, visitado, componente_actual):
        visitado[inicio] = True
        componente_actual[0] += 1

        for vecino in self.vertices[inicio]:
            if not visitado[vecino]:
                self.dfs(vecino, visitado, componente_actual)

    def encontrar_componentes_conexos(self):
        visitado = {vertice: False for vertice in self.vertices}
        tamaños_componentes = []

        for vertice in self.vertices:
            if not visitado[vertice]:
                componente_actual = [0]
                self.dfs(vertice, visitado, componente_actual)
                tamaños_componentes.append(componente_actual[0])

        return tamaños_componentes

def main():
    grafo = Grafo()
    r=int(input("Cantidad de conexiones: "))
    for i in range(r):
        t=input("Ingrese conexión: ")
        p=t.split()
        grafo.agregar_arista(int(p[0]), int(p[1]))

    tamaños_componentes = grafo.encontrar_componentes_conexos()

    if tamaños_componentes:
        min_size = min(tamaños_componentes)
        max_size = max(tamaños_componentes)

        print(min_size,max_size)
    else:
        print("El grafo no tiene vértices")

if __name__ == "__main__":
    main()
```
