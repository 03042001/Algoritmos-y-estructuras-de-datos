# Requisitos

# Especificación
En la hamburguesería de Jim hay una fila de gente esperando a ser atendidos, a cada cliente se le entrega un número de orden y cada plato tiene un tiempo de preparación, la idea es organizar a los clientes del primero en ser atendido, es decir, cuyo plato está listo en el menor tiempo posible, al ultimo cliente en ser atendido; el tiempo de demora es la suma entre el numero de orden y el tiempo de preparación del plato.

# Entrada
El numero n de clientes, seguido por n lineas de pares de enteros separados por espacio.


# Salida
Una lista con los clientes organizados desde el primero en ser atendido hasta el último

# Diseño

---



# Estrategia
Vamos a tomar los datos de la entrada como una matriz de n * 2, sumamos cada tupla, la cual equivale a cada elemento de la matriz, y esto nos da una lista, la cual vamos a ordenar mediante el método de insertion sort, para luego dar el resultado del ejercicio

# Estructura de datos
En este ejercicio usaremos listas(arrays) como estructura de datos para almacenar y manipular la información.

# Implementación del Algoritmo



---



```python
def insertion_sort(list1, list2):
    """Usamos insertion sort para organizar ambas listas al mismo tiempo
    (list, list) -> list
    """
    for j in range(1, len(list1)):
        key = list1[j]
        key2 = list2[j]
        i = j - 1

        while i >= 0 and list1[i] > key:
            list1[i + 1] = list1[i]
            list2[i + 1] = list2[i]
            i = i - 1
        list1[i + 1] = key
        list2[i + 1] = key2

    return list2

def jimOrders(orders):
     """Hacemos la suma del numero de orden y del tiempo de preparación
      y llamamos a la función insertion sort para entregar el resultado
     list -> list
     """
     lista = []
     numbers = []
     for i in range(0, len(orders)):
        suma = orders[i][0] + orders[i][1]
        lista.append(suma)

     cont = 1
     while len(numbers) < len(lista):
          numbers.append(cont)
          cont += 1

     return insertion_sort(lista, numbers)

def main():
    n = int(input().strip())

    orders = []

    for _ in range(n):
        orders.append(list(map(int, input().rstrip().split())))

    result = jimOrders(orders)
    print(*result)

main()
```

# Invariante
El invariante lo analizaremos principalmente en la implementación del método de ordenamiento "insertion sort", en este podemos observar que al finalizar cada iteración del bucle externo, los elementos hasta la posición actual están ordenados.
