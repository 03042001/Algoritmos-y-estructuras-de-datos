# Estrategia

Para solucionar este ejercicio lo que haremos será crear una clase libro que tenga un genero, un autor y un titulo, cada libro que agreguemos tendrá estos 3 atributos, pero ademas crearemos otra clase biblioteca en la que guardaremos los libros, crearemos una funcion para agregar libros en la cual llamaremos a la clase libros, y la otra funcion será para organizar la biblioteca en la cual iremos imprimiendo mediante varios ciclos cada libro, esto lo hacemos porque un genero puede tener muchos autores, y estos a su vez pueden tener mas de un libro.


# Estructura de datos

Para este ejercicio no me parecio practico usar la estructura de datos de arboles, esto debido a que no entendia bien como seria la insercion de datos a un lado u otro del arbol, sin embargo si sería bastante funcional si cada raiz de cada arbol fuera un genero y por lo tanto se crearia un arbol por genero literario, cada uno con sus autores y estos a su vez con sus libros, sin embargo no logre implementar el ejercicio de esa forma y opte por resolverlo como pude debido al tiempo.

# Implementación del Algoritmo

```python

class Libro:
    def __init__(self, genero, autor, titulo):
        self.genero = genero
        self.autor = autor
        self.titulo = titulo

    def __str__(self):
        return f'Género: {self.genero}, Autor: {self.autor}, Título: "{self.titulo}"'

class Biblioteca:
    def __init__(self):
        self.biblioteca = {}

    def agregar_libro(self, libro):
        genero = libro.genero
        autor = libro.autor

        if genero not in self.biblioteca:
            self.biblioteca[genero] = {}

        if autor not in self.biblioteca[genero]:
            self.biblioteca[genero][autor] = []

        self.biblioteca[genero][autor].append(libro)

    def organizar_biblioteca(self):
        print("Biblioteca Organizada:")
        for genero, autores in self.biblioteca.items():
            print(f'- {genero}')
            for autor, libros in autores.items():
                print(f'- Autor: {autor}')
                for libro in libros:
                    print(f'- {libro}')

def main():
    n = int(input("Número de libros a agregar: "))
    biblioteca = Biblioteca()

    for _ in range(n):
        genero = input("Género: ")
        autor = input("Autor: ")
        titulo = input("Título: ")
        libro = Libro(genero, autor, titulo)
        biblioteca.agregar_libro(libro)

    biblioteca.organizar_biblioteca()

if __name__ == "__main__":
    main()
```
