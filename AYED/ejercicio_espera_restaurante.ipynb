# Especificación

Tieu owns a pizza restaurant and he manages it in his own way. While in a normal restaurant, a customer is served by following the first-come, first-served rule, Tieu simply minimizes the average waiting time of his customers. So he gets to decide who is served first, regardless of how sooner or later a person comes.

Different kinds of pizzas take different amounts of time to cook. Also, once he starts cooking a pizza, he cannot cook another pizza until the first pizza is completely cooked. Let's say we have three customers who come at time t=0, t=1, & t=2 respectively, and the time needed to cook their pizzas is 3, 9, & 6 respectively. If Tieu applies first-come, first-served rule, then the waiting time of three customers is 3, 11, & 16 respectively. The average waiting time in this case is (3 + 11 + 16) / 3 = 10. This is not an optimized solution. After serving the first customer at time t=3, Tieu can choose to serve the third customer. In that case, the waiting time will be 3, 7, & 17 respectively. Hence the average waiting time is (3 + 7 + 17) / 3 = 9.

Help Tieu achieve the minimum average waiting time. For the sake of simplicity, just find the integer part of the minimum average waiting time.

# Entrada
The first line contains an integer N, which is the number of customers.
In the next N lines, the ith line contains two space separated numbers Ti and Li. Ti is the time when ith customer order a pizza, and Li is the time required to cook that pizza.

The  customer i is not the customer arriving at the i arrival time.

# Salida

Display the integer part of the minimum average waiting time.

# Estrategia
Crearemos un bucle que se ejecutará mientras haya pedidos pendientes, dentro de este llamaremos a una funcion que es la que se encarga de procesar los pedidos, lo que esta hace es que mientras haya pedidos en la lista principal  cuyo tiempo sea menor o igual al tiempo actual, los traslada a la cola de pedidos pendientes. Luego, si hay pedidos pendientes, selecciona el que tiene el tiempo de procesamiento mínimo y actualiza el tiempo de espera total y el tiempo actual.  Necesitaremos crear algunas funciones auxiliares para tareas  como agregar pedidos, saber si hay pedidos pendientes y una que nos entregue el pedido con el tiempo de procesamiento minimo. Finalmente, imprimimos el tiempo promedio de espera dividiendo el tiempo total de espera entre el número total de pedidos, convirtiéndolo a entero

# Estructura de datos
Para este ejercicio usaremos una estructura de datos parecida a una cola, pero en el contructor agregaremos tambien un tiempo de espera y  un tiempo actual, estos los inicializaremos en 0.

# Implementación del Algoritmo



---


```python
class ProcesadorPedido:
    def __init__(self, pedidos):
        self.ttespera = 0
        self.npedidos = len(pedidos)
        self.ppendientes = []
        self.tactual = 0
        self.pedidos = pedidos

    def procesar_pedidos(self):
        while self.hpedidos():
            self.ptactual()

    def hpedidos(self):
        return len(self.ppendientes) > 0 or len(self.pedidos) > 0

    def ptactual(self):
        while len(self.pedidos) > 0 and self.pedidos[0][0] <= self.tactual:
            pedido = self.tpedido(self.pedidos)
            self.agregar_pedido(self.ppendientes, (pedido[1], pedido[0]))

        if len(self.ppendientes) > 0:
            mepedido = self.tpedido(self.ppendientes)
            tespera = self.tactual - mepedido[1] + mepedido[0]
            self.ttespera += tespera
            self.tactual += mepedido[0]
        else:
            self.tactual += 1

    def tpedido(self, lista_pedidos):
        iminimo = 0
        for i in range(1, len(lista_pedidos)):
            if lista_pedidos[i][0] < lista_pedidos[iminimo][0]:
                iminimo = i
        return lista_pedidos.pop(iminimo)

    def agregar_pedido(self, lista_pedidos, pedido):
        lista_pedidos.append(pedido)

if __name__ == "__main__":
    n = int(input())
    pedidos = []

    for i in range(n):
        linea = input().split()
        l, t = int(linea[0]), int(linea[1])
        pedidos.append((l, t))

    procesador_pedido = ProcesadorPedido(pedidos)
    procesador_pedido.procesar_pedidos()
    print(int(procesador_pedido.ttespera / procesador_pedido.npedidos))
```
