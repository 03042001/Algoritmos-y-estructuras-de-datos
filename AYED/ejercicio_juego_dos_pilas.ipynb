Ejercicio: "Game of two stacks"

# Requisitos

---



# Especificación
Lo que tenemos que hacer es determinar la máxima cantidad de numeros que puede retirar Nick de las dos pilas sin exceder la suma máxima permitida.


# Entrada
La entrada consiste en: Una entero que será la suma máxima permitida y dos arreglos, el arreglo 'a' representa la primera pila de numeros y 'b' la segunda pila.

# Salida
Un entero que corresponde al maximo numero de selecciones que Nick puede hacer.


# Diseño

---



# Estrategia
Para solucionar el ejercicio usaremos dos contadores, uno que se encargará del numero de formas que llevamos y el otro en el cual vamos sumando los elementos para verificar que no se pase de la suma maxima permitida mediante un ciclo, dentro de este ciclo usaremos condicionales para los diferentes casos que nos podemos encontrar: que a sea vacio, que b sea vacio, que a en la posicion 0 sea menor que b en esa misma posicion o el resto de casos. A medida que se realicen estos condicionales se quitará de una de las pilas (dependiendo del caso) y ese valor se sumará al contador suma, asi mismo se sumará en contador de veces 1 unidad, al final retornamos este contador como resultado.


# Estructura de datos
Para la solucion de este ejercicio se usa una estructura lineal de datos llamada pila, esto debido a la forma en que esta planteado el ejercicio.


# Implementación del Algoritmo



---


```python
def twoStacks(maxSum, a, b):
    suma = 0
    cont = 0
    while suma <= maxSum and (a or b):
        if not a:
            suma += b.pop(0)
        elif not b:
            suma += a.pop(0)
        elif a[0] < b[0]:
            suma += a.pop(0)
        else:
            suma += b.pop(0)
        cont += 1
    return cont - 1
```

# Invariante

El invariante podría estar asociado a la forma en que se trabaja con las pilas, es decir que tanto en la inicializacion, estabilidad y fin del algoritmo siempre el ultimo en entrar a la pila es el primero en salir de esta.
