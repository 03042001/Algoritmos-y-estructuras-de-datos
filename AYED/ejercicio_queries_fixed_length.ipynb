# Especificación
Considerando una secuencia de enteros, queremos calcular el resultado de:

(min(max aj))

Dado un arreglo y q consultas retornar el resultado de las consultas.


# Entrada
Dos enteros separados por espacio n y q

n enteros separados por espacio, los elementos del arreglo.

un entero denotando el valor de d para la consulta.

# Salida
la respuesta a cada consulta


# Diseño

---



# Estrategia
Lo que haremos será iterar sobre el rango de las consultas, iremos sacando los maximos mediante ciclos y eliminando de la cola los elementos que no nos sirvan, al final nos quedará una lista con los maximos y lo que haremos será entregar el minimo de estos que será nuestro resultado.

# Estructura de datos
Para este ejercicio haremos una estructura de datos que tendrá en el constructor un arrelglo, y dos valores que serán n y q, será parecido a una cola.

# Implementación del Algoritmo



---


```python
class Querie:
    def __init__(self, n, q, arr):
        self.n = n
        self.q = q
        self.arr = arr

    def proceso(self):
        for l in range(self.q):
            d = int(input().strip())
            maximos = []
            cola = []

            for i in range(self.n):
                while cola and cola[0] < i-d+1:
                    cola.pop(0)

                while cola and self.arr[i] >= self.arr[cola[-1]]:
                    cola.pop()

                cola.append(i)

                if i >= d - 1:
                    maximos.append(self.arr[cola[0]])

            print(min(maximos))


if __name__ == "__main__":
    n, q = [int(x) for x in input().strip().split()]
    arr = [int(x) for x in input().strip().split()]

    maximo_consulta_colas = Querie(n, q, arr)
    maximo_consulta_colas.proceso()
```
