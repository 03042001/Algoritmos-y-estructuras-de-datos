# Requisitos

---



# Especificación

Dado un entero x necesitamos encontrar el super digito del entero, si x tiene solo un dígito el super digito es x, y si no el super digito es igual al super digito de la suma de los digitos de x.


# Entrada

La entrada consiste en dos enteros, n y k, separados por espacio donde n es el entero al que le encontramos el super digito y k es el numero de veces que concatenamos n.

# Salida

Un entero x que es el super digito que debiamos calcular.

# Diseño

---



# Estrategia

Lo primero que vamos a hacer es separar el numero por digitos dentro de una lista, esta misma lista la vamos a concatenar k veces, posteriormente sumamos cada elemento de la lista, es decir cada digito del numero, el resultado lo volvemos a dividir en digitos y si es un número de dos digitos o más volvemos a ejecutar la función recursivamente hasta que sea de longitud 1, este será nuestro resultado.

# Estructura de datos

Utilizaremos listas.


# Implementación del Algoritmo



---


```python
def superDigit(n, k, dinamica= {}):
    if (n, k) in dinamica:
      return dinamica[(n, k)]
    x = [int(a) for a in str(n)]
    if len(x) == 1:
        return x[0]
    num = x * k
    suma = sum(num)
    dinamica[(n, k)] = suma
    return superDigit(suma, 1, dinamica)

def main():
    first_multiple_input = input().rstrip().split()
    n = first_multiple_input[0]
    k = int(first_multiple_input[1])
    result = superDigit(n, k)
    print(result)

main()
```

# Invariante

No hay un invariante especifico en este algoritmo.
