# Especificación
Debes decidir, dada una paleta de exclusivamente dos colores Rojo y Azul, si un grafo
fuertemente conexo G puede ser coloreado, de tal manera que ninguno de los nodos
conectados tenga el mismo color con alguno de sus vecinos.
Se puede suponer que el grafo G no contiene ciclos y es no dirigido ( Para todo arco
(Ai
, Bi), existe el arco (Bi
, Ai) ).



# Entrada

La entrada consiste en varios casos de prueba. Cada caso comienza con un número
n(1 ≤ n ≤ 20), denotando la cantidad de nodos de G. La segunda línea contiene el
número de arcos E. Seguido ingresarán E líneas con parejas x y y, denotando los arcos
(Ai
, Bi).
La entrada termina con una línea donde n = 0 y no debe ser procesada.

# Salida

Debe imprimir BICOLORABLE. , en caso de que G, pueda ser pintado con los dos
colores cumpliendo con las restricciones y NOT BICOLORABLE. , en caso contrario.

# Estrategia

Lo que haremos será verificar la conectividad del grafo y luego intentar colorear los vértices de manera que ningún par de vértices tenga el mismo color, para esto usaremos el código de grafo que ya tenemos pero le haremos algunas modificaciones.

# Estructura de datos
Para este ejercicio implementaremos un grafo.

# Implementación del Algoritmo



---


```python
from typing import List

class Graph:
    def __init__(self, vertices: int):
        self.vertices = vertices
        self.graph = [[0] * vertices for _ in range(vertices)]

    def add_edge(self, u: int, v: int):
        self.graph[u][v] = 1

    def is_connected(self) -> bool:
        visited = [False] * self.vertices
        queue = [0]
        visited[0] = True

        while queue:
            node = queue.pop(0)
            for i in range(self.vertices):
                if self.graph[node][i] == 1 and not visited[i]:
                    queue.append(i)
                    visited[i] = True

        return all(visited)

    def solve(self) -> str:
        if not self.is_connected():
            return "NOT BICOLORABLE."

        color = [-1] * self.vertices

        def is_bicolorable_util(v: int, color: List[int]) -> bool:
            for i in range(self.vertices):
                if self.graph[v][i] == 1:
                    if color[i] == -1:
                        color[i] = 1 - color[v]
                        if not is_bicolorable_util(i, color):
                            return False
                    elif color[i] == color[v]:
                        return False
            return True

        for i in range(self.vertices):
            if color[i] == -1:
                color[i] = 1
            if not is_bicolorable_util(i, color):
                return "NOT BICOLORABLE."

        return "BICOLORABLE."

def main():
    case_number = 1
    while True:
        n = int(input())
        m = int(input())

        if n == 0:
            break

        graph = Graph(n)

        for i in range(m):
            a, b = map(int, input().split())
            graph.add_edge(a, b)

        result = graph.solve()
        print(f"{result}")
        case_number += 1

if __name__ == "__main__":
    main()
```
