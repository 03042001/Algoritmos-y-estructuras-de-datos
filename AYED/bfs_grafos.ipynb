# Estrategia
Para este ejercicio usaremos la misma estructura del ejercicio pasado de las listas y matrices de adyacencia, e implementaremos la busqueda por profundidad  y por anchura, para la busqueda por profundidad seleccionamos un vertice inicial y lo marcamos como visitado, despues exploramos uno de los vertices vecinos no visitados, cuando no haya mas vecinos no visitados vamos al vertice anterior y continuamos explorando los no visitados, y repetimos estos pasos hasta haber visitado todos los nodos o vertices alcanzables desde el inicial.  Para la busqueda por amplitud lo que haremos será tomar un nodo inicial y marcarlo como visitado, agregamos este nodo inicial a una cola, mientras la cola no esté vacia saca un nodo de la cola y lo visita y agregamos todos los vecinos de este nodo no visitados a la cola y los marcamos como visitados.

# Estructura de datos

En este ejercicio usamos los grafos como estructura de datos.

# Implementación del Algoritmo



---




```python
class Grafo:
    def __init__(self, num_vertices):
        self.num_vertices = num_vertices
        self.lista_adyacencia = [[] for _ in range(num_vertices)]
        self.matriz_adyacencia = [[0] * num_vertices for _ in range(num_vertices)]

    def agregar_arista(self, vertice1, vertice2):
        self.lista_adyacencia[vertice1].append(vertice2)
        self.lista_adyacencia[vertice2].append(vertice1)
        self.matriz_adyacencia[vertice1][vertice2] = 1
        self.matriz_adyacencia[vertice2][vertice1] = 1

    def dfs_lista(self, vertice, visitados):
        visitados[vertice] = True
        print(vertice, end=" ")

        for adyacente in self.lista_adyacencia[vertice]:
            if not visitados[adyacente]:
                self.dfs_lista(adyacente, visitados)

    def dfs_matriz(self, vertice, visitados):
        visitados[vertice] = True
        print(vertice, end=" ")

        for i in range(self.num_vertices):
            if self.matriz_adyacencia[vertice][i] == 1 and not visitados[i]:
                self.dfs_matriz(i, visitados)

    def bfs_lista(self, vertice):
        visitados = [False] * self.num_vertices
        cola = []
        visitados[vertice] = True
        cola.append(vertice)

        while cola:
            vertice = cola.pop(0)
            print(vertice, end=" ")

            for adyacente in self.lista_adyacencia[vertice]:
                if not visitados[adyacente]:
                    visitados[adyacente] = True
                    cola.append(adyacente)

    def bfs_matriz(self, vertice):
        visitados = [False] * self.num_vertices
        cola = []
        visitados[vertice] = True
        cola.append(vertice)

        while cola:
            vertice = cola.pop(0)
            print(vertice, end=" ")

            for i in range(self.num_vertices):
                if self.matriz_adyacencia[vertice][i] == 1 and not visitados[i]:
                    visitados[i] = True
                    cola.append(i)


```
