**1 Ejercicio Palíndromo**
-Diseñar e implementar un algoritmo que verifique si una cadena o grupo de cadenas es un palíndromo. Las cadenas se separan por espacio. eje1: "abcdcba 12321"

```python
def palindrome(lista):
  """
  Funcion que verifica si todos los elementos de una lista son palindromos
  list -> Boolean
  """
  nueva = []
  for i in range(0, len(lista)):
    reves = lista[i][::-1]
    nueva.append(reves)
  if lista == nueva:
    return True
  else:
    return False

def main():
  while 0 < 10:
    entrada = input()
    lista = entrada.split(' ')
    print(palindrome(lista))

main()
```

**2 Problema de Cadenas Rotadas:**
-Dadas dos cadenas "cadena1" y "cadena2", implementa un algoritmo que determine si la "cadena2" es una rotación cíclica de "cadena1". Por ejemplo, si "cadena1" es "abcde" y "cadena2" es "deabcd", entonces deberías retornar True.

```python
def cadenas_rotadas(cadena1, cadena2):
  """
  Verifica si la cadena más pequeña se encuentra identica en la cadena más grande
  (str, str)-> Boolean
  """
  if cadena1 < cadena2:
      cadena1 = cadena1 + cadena1
  else:
    cadena2 = cadena2 + cadena2

  return cadena2 in cadena1

def main():
  cadena1 = input()
  cadena2 = input()
  print(cadenas_rotadas(cadena1, cadena2))

main()
```

**3 Problema de Subcadenas Distintas:**
-Dada una cadena, encuentra la cantidad de subcadenas distintas que se pueden formar utilizando sus caracteres e imprimalas.

```python
def distintas(input_str):
    """Encuentra las distintas subcadenas
    str -> set
    """
    subcadenas_distintas = set()

    n = len(input_str)
    for i in range(n):
        for j in range(i + 1, n + 1):
            substring = input_str[i:j]
            subcadenas_distintas.add(substring)

    return subcadenas_distintas

input_str = input("Ingresa una cadena: ")
substrings = distintas(input_str)

print(f"Cantidad de subcadenas distintas: {len(substrings)}")
for substring in substrings:
    print(substring)
```

**4 Problema de Números Primos Gemelos:**
-Encuentra todos los pares de números primos gemelos dentro de un rango dado. Los números primos gemelos son dos números primos consecutivos que difieren en 2.

```python
def primo(n):
    """Encuentra si un numero es primo o no
    int -> Boolean
    """
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

def primos_gemelos(inicio, fin):
    pares_gemelos = []
    for num in range(inicio, fin - 1):
        if primo(num) and primo(num + 2):
            pares_gemelos.append((num, num + 2))
    return pares_gemelos

inicio = int(input("Inicio: "))
fin = int(input("Fin: "))

pares_primos_gemelos = primos_gemelos(inicio, fin)
print(pares_primos_gemelos)
```
