# Estrategia

Para este ejercicio tomaremos la estructura del grafo, en el constructor estableceremos cada vértice como un diccionario, esto nos facilitará las cosas cuando hagamos la DFS y le pongamos el estado de visitado a los vértices por los que ya pasamos; además de las funciones básicas implementaremos una busqueda por profundidad como dijimos anteriormente para ir explorando el grafo e ir calculando los tamaños de todos los componentes conexos, esto lo guardaremos en una lista que retornaremos. Por último dentro de la función main sacaremos el mínimo y el máximo de esa lista de tamaños que retornamos anteriormente y estos dos enteros serán el resultado que devolverá el algoritmo.

# Estructura de datos

Usaremos un grafo como estructura de datos, cada vértice lo tomaremos como un diccionario. Usaremos además la busqueda por profundidad.

# Implementación del Algoritmo



---


```python
class Grafo:
    def __init__(self):
        self.vertices = {}

    def agregar_vertice(self, vertice):
        if vertice not in self.vertices:
            self.vertices[vertice] = []

    def agregar_arista(self, vertice1, vertice2):
        self.agregar_vertice(vertice1)
        self.agregar_vertice(vertice2)

        self.vertices[vertice1].append(vertice2)
        self.vertices[vertice2].append(vertice1)

    def dfs(self, inicio, visitado, componente_actual):
        visitado[inicio] = True
        componente_actual[0] += 1

        for vecino in self.vertices[inicio]:
            if not visitado[vecino]:
                self.dfs(vecino, visitado, componente_actual)

    def encontrar_componentes_conexos(self):
        visitado = {vertice: False for vertice in self.vertices}
        tamaños_componentes = []

        for vertice in self.vertices:
            if not visitado[vertice]:
                componente_actual = [0]
                self.dfs(vertice, visitado, componente_actual)
                tamaños_componentes.append(componente_actual[0])

        return tamaños_componentes

def main():
    grafo = Grafo()
    r=int(input(""))
    for i in range(r):
        t=input("")
        p=t.split()
        grafo.agregar_arista(int(p[0]), int(p[1]))

    tamaños_componentes = grafo.encontrar_componentes_conexos()

    if tamaños_componentes:
        min_size = min(tamaños_componentes)
        max_size = max(tamaños_componentes)

        print(min_size,max_size)
    else:
        print("El grafo no tiene vértices")

if __name__ == "__main__":
    main()
```
