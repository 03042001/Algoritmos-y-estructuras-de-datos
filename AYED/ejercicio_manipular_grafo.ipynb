# Estrategia
En este ejercicio en el constructor vamos a establecer un numero de vertices que tendrá el grafo asi como una lista de adyacencia y una matriz de adyacencia, haremos tambien una funcion para agregar aristas, esta funcion es importante porque es la encargada de modificar tanto la lista de adyacencias como la matriz de adyacencia conforme a las aristas que haya en el grafo, habria que crear tambien unas funciones para imprimir tanto la lista como la matriz de adyacencia, estas funciones usarán ciclos para ir imprimiendo los diferentes elementos.


# Estructura de datos

En este ejercicio usamos los grafos como estructura de datos.

# Implementación del Algoritmo



---




```python
class Grafo:
    def __init__(self, num_vertices):
        self.num_vertices = num_vertices
        self.lista_adyacencia = [[] for _ in range(num_vertices)]
        self.matriz_adyacencia = [[0] * num_vertices for _ in range(num_vertices)]

    def agregar_arista(self, vertice1, vertice2):
        self.lista_adyacencia[vertice1].append(vertice2)
        self.lista_adyacencia[vertice2].append(vertice1)
        self.matriz_adyacencia[vertice1][vertice2] = 1
        self.matriz_adyacencia[vertice2][vertice1] = 1

    def imprimir_lista_adyacencia(self):
        for i, adyacentes in enumerate(self.lista_adyacencia):
            print(f"Vértice {i}: {adyacentes}")

    def imprimir_matriz_adyacencia(self):
        for fila in self.matriz_adyacencia:
            print(f"{fila}")

def calcular_e_imprimir(grafo):
    print("Lista de Adyacencia:")
    grafo.imprimir_lista_adyacencia()

    print("\nMatriz de Adyacencia:")
    grafo.imprimir_matriz_adyacencia()
```

# Casos de prueba


num_vertices = 4


mi_grafo = Grafo(num_vertices)


mi_grafo.agregar_arista(0, 1)


mi_grafo.agregar_arista(0, 2)


mi_grafo.agregar_arista(1, 3)

calcular_e_imprimir(mi_grafo)
