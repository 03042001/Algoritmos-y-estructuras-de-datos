# Requisitos

---



# Especificación

Dado un numero de dolares y un arreglo con las denominaciones que se pueden usar, determinar cuantas maneras hay para cambiar el dinero.

# Entrada
La primera linea de la entrada consiste en dos numeros enteros, n y m, la cantidad a ser cambiada y el numero de denominaciones; la segunda linea contiene m numero de enteros separados por espacio.

# Salida

La salida consiste en un entero que es la cantidad de maneras posibles que hay para cambiar el dinero.

# Diseño

---



# Estrategia
Para solucionar el problema usaremos un contador que nos ayudará a la hora de ir recorriendo el arreglo y tambien a la hora de guardar los resultados mediante programación dinámica; lo que haremos será ir calculando el numero de formas que hay de cambiar el dinero que tenemos pero lo haremos con la denominacion que tenemos en el momento debido al contador y despues el numero de formas sin usar esa denominacion, y esto lo hacemos recurrentemente hasta llegar a uno de los casos base.

# Implementación del Algoritmo



---


```python
def ways(n, coins, index=0, dinamica={}):
    """Devuelve el numero de formas de cambiar cierta cantidad de dinero con
    unas denominaciones dadas
    """
    if n == 0:
        return 1
    if n < 0 or index == len(coins):
        return 0
    if (n, index) in dinamica:
        return dinamica[(n, index)]

    con_denom_actual = ways(n - coins[index], coins, index, dinamica)
    sin_denom_actual = ways(n, coins, index + 1, dinamica)
    total = con_denom_actual + sin_denom_actual
    dinamica[(n, index)] = total

    return total

def main():
    first_multiple_input = input().rstrip().split()
    n = int(first_multiple_input[0])
    m = int(first_multiple_input[1])
    coins = list(map(int, input().rstrip().split()))
    res = ways(n, coins)
    return res

main()
```
