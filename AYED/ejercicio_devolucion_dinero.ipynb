```python
def ways(n, coins, index=0, dinamica={}):
    """Devuelve el numero de formas de cambiar cierta cantidad de dinero con
    unas denominaciones dadas
    """
    if n == 0:
        return 1
    if n < 0 or index == len(coins):
        return 0
    if (n, index) in dinamica:
        return dinamica[(n, index)]

    con_denom_actual = ways(n - coins[index], coins, index, dinamica)
    sin_denom_actual = ways(n, coins, index + 1, dinamica)
    total = con_denom_actual + sin_denom_actual
    dinamica[(n, index)] = total

    return total

def main():
    first_multiple_input = input().rstrip().split()
    n = int(first_multiple_input[0])
    m = int(first_multiple_input[1])
    coins = list(map(int, input().rstrip().split()))
    res = ways(n, coins)
    return res

main()
```
