# Estrategia

Lo que hacemos primero es crear el constructor __init__ que nos ayudará a inicializar los atributos necesarios para implementar un grafo, lo principal que necesitamos para el grafo son los vertices que en este caso guardaremos en un set, una vez hecho esto podemos comenzar a crear distintas funciones como agregar vertices, funcion en la cual verificamos si el vertice que queremos agregar no está presente en la lista de vertices del grafo, en este caso se agrega a la lista de vertices, se crea una entrada en el diccionario 'self.vertices' para el nuevo vertice y se inicializa con una lista de adyacencias.  Esta lista de adyacencias nos servirá a la hora de agregar las aristas, que son las conexiones entre los diferentes vertices. Y así iremos creando diferentes funciones que se ajusten al objetivo que tenemos al implementar el grafo.


# Estructura de datos

En este ejercicio implementaremos la clase responsable para modelar un grafo.

# Implementación del Algoritmo



---




```python
class Grafo:
    def __init__(self):
        self.vertices = {}

    def agregar_vertice(self, vertice):
        if vertice not in self.vertices:
            self.vertices[vertice] = []

    def agregar_arista(self, vertice1, vertice2):
        if vertice1 in self.vertices and vertice2 in self.vertices:
            self.vertices[vertice1].append(vertice2)
            self.vertices[vertice2].append(vertice1)

    def eliminar_vertice(self, vertice):
        if vertice in self.vertices:
            del self.vertices[vertice]
            for v in self.vertices:
                self.vertices[v] = [x for x in self.vertices[v] if x != vertice]

    def buscar_vertice(self, vertice):
        return vertice in self.vertices

    def buscar_arista(self, vertice1, vertice2):
        return vertice1 in self.vertices and vertice2 in self.vertices and vertice2 in self.vertices[vertice1]

    def obtener_conexiones(self, vertice):
        if vertice in self.vertices:
            return self.vertices[vertice]
        return []

    def __str__(self):
        grafo_str = ""
        for vertice, conexiones in self.vertices.items():
            grafo_str += f"{vertice}: {', '.join(conexiones)}\n"
        return grafo_str

```
