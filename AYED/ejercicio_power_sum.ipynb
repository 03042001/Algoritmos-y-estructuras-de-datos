# Examen punto 1
Ejercicio: "The power sum"

# Requisitos

---



# Especificación

Encontrar el numero de formas que un entero dado x, se puede expresar como la suma de las potencias de números naturales únicos.

# Entrada

La entrada corresponde a una serie consecutiva de casos, cada caso corresponde a dos lineas de entrada, con enteros x y n respectivamente.

# Salida


Imprima el numero de posibles combinaciones para secuencias de numeros naturales unicos elevados a la enesima potencia que suman x.

# Diseño

---



# Estrategia

Dentro de la función powerSum, calculamos la diferencia entre x y num elevado a la potencia n. Si esta diferencia es menor que 0, significa que hemos superado x, por lo que retornamos 0.

Si la diferencia es igual a 0, significa que hemos encontrado una combinación de números que suma exactamente x, por lo que retornamos 1.

Si la diferencia es mayor que 0, entonces significa que a esa diferencia le podemos seguir buscando números que elevados a n nos den el valor deseado x (esta busqueda de los números la hacemos mediante el index "num") pero haremos la llamada recursiva dos veces, una con el valor de la diferencia que dió mayor a 0  y al index le sumamos 1 para continuar con esta busqueda, y otra llamada recursiva con el numero x original pero le sumamos al index "num" una unidad. lo unico que faltaría por hacer es sumar los resultados de ambas llamadas recursivas y retornar ese valor como el número de formas en que x se puede expresar como la suma de números elevados a la potencia n.


# Implementación del Algoritmo



---


```python
def powerSum(x, n, num=1):
    diff = x - num**n
    if diff < 0:
        return 0
    elif diff == 0:
        return 1
    else:
        return (powerSum(x, n, num + 1) +
                powerSum(diff, n, num + 1))

def main():
    X = int(input().strip())
    N = int(input().strip())
    result = powerSum(X, N)
    print(result)
main()
```

# Invariante

En el algoritmo el index "num" va a estar aumentando constantemente, además el resultado nunca será negativo.

1.   Iniciación: El index "num" empieza en 1, y el resultado mínimo es 0 o positivo.
2.   Estabilidad: Durante el proceso num va aumentando constantemente y el resultado es positivo.
3. Terminación: Cuando llegue a su recursión maxima (casos base), el resultado permanecerá mayor que 0 y el indice num será positivo.
