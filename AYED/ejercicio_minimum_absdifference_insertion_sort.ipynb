Ejercicio #2: "Minimum absolute difference in an array"


# Requisitos

---



# Especificación
Tenemos que calcular la diferencia mínima absoluta entre un par de números dentro de un arreglo dado.

# Entrada
Lo primero en la entrada será un entero n, que será el tamaño del arreglo y posteriormente n enteros separados por espacio.

# Salida
La salida es unicamente un entero n que será la diferencia mínima absoluta encontrada por el programa.

# Diseño

---



# Estrategia
Lo primero que vamos a hacer con el arreglo que nos den será organizarlo, con esto haremos que los elementos cercanos estén agrupados y nos quedará más sencillo encontrar la diferencia mínima entre elementos adyacentes; posteriormente creamos una lista nueva en la que guardaremos todas las diferencias, las cuales calcularemos mediante un ciclo que recorra todo el arreglo original, lo último que quedaría por hacer es sacarle valor absoluto a cada valor de esta lista, lo podriamos hacer mediante abs(), pero para hacerlo manual lo que hacemos es multiplicar cada resultado por -1, y retornamos el mínimo de la lista de diferencias.

# Estructura de datos
En este ejercicio usaremos principalmente listas(arrays).

# Implementación del Algoritmo



---



```python
def insertion_sort(arr):
    """Método de organización insertion sort
    list -> list
    """
    for j in range(1, len(arr)):
        key = arr[j]
        i = j - 1

        while i >= 0 and arr[i] > key:
            arr[i + 1] = arr[i]
            i = i - 1
        arr[i + 1] = key

    return arr

def minimumAbsoluteDifference(arr):
    """Encuentra la diferencia mínima absoluta del arreglo
    list -> int
    """
    sorted_arr = insertion_sort(arr)
    diff = []
    for i in range(1, len(arr)):
        difference = sorted_arr[i] - sorted_arr[i - 1]
        if difference < 0:
            difference = -difference

        diff.append(difference)
    return min(diff)

def main():
    n = int(input().strip())
    arr = list(map(int, input().rstrip().split()))
    result = minimumAbsoluteDifference(arr)
    print(result)

main()
```

# Invariante
Después de cada iteración del bucle for, la lista diff contendrá las diferencias absolutas entre elementos adyacentes en la parte ordenada del arreglo. Estas diferencias estarán en orden ascendente.
